```python
import os
import numpy as np
import pywt
import cv2
from skimage.feature import graycomatrix, graycoprops
import pandas as pd
```

```python
def extract_features_from_images(image_files, labels):
    features_list = []

    for img_path in image_files:
        img = cv2.imread(img_path)
        coeffs2 = pywt.dwt2(img, 'db2')
        LL1, (LH1, HL1, HH1) = coeffs2
        coeffs3 = pywt.dwt2(LL1, 'db2')
        LL2, (LH2, HL2, HH2) = coeffs3
        coeffs4 = pywt.dwt2(LL2, 'db2')
        LL3, (LH3, HL3, HH3) = coeffs4
        coeffs5 = pywt.dwt2(LL3, 'db2')
        LL4, (LH4, HL4, HH4) = coeffs5

        
        # Extract features
        g = []
        
        g.append(graycomatrix(np.uint8(LL1[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL1[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL1[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH1[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH1[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH1[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL1[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL1[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL1[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH1[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH1[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH1[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        
        g.append(graycomatrix(np.uint8(LL2[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL2[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL2[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH2[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH2[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH2[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL2[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL2[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL2[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH2[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH2[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH2[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        
        g.append(graycomatrix(np.uint8(LL3[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL3[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL3[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH3[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH3[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH3[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL3[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL3[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL3[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH3[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH3[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH3[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        
        g.append(graycomatrix(np.uint8(LL4[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL4[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LL4[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH4[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH4[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(LH4[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL4[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL4[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HL4[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH4[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH4[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
        g.append(graycomatrix(np.uint8(HH4[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        
#         g.append(graycomatrix(np.uint8(LL5[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LL5[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LL5[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LH5[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LH5[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LH5[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HL5[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HL5[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HL5[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HH5[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HH5[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HH5[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
        
#         g.append(graycomatrix(np.uint8(LL6[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LL6[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LL6[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LH6[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LH6[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(LH6[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HL6[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HL6[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HL6[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HH6[:,:,0]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HH6[:,:,1]), [1], [0], levels=256, normed= True, symmetric=True))
#         g.append(graycomatrix(np.uint8(HH6[:,:,2]), [1], [0], levels=256, normed= True, symmetric=True))

        con = []
        enr = []
        dis = []
        hom = []
        for t in range(0, len(g)):
            con.append(graycoprops(np.array(g[t]), 'contrast'))
            enr.append(graycoprops(np.array(g[t]), 'energy'))
            dis.append(graycoprops(np.array(g[t]), 'dissimilarity'))
            hom.append(graycoprops(np.array(g[t]), 'homogeneity'))

        # Concatenate features
        con_features = np.reshape(np.array(con).ravel(), (1, len(np.array(con).ravel())))
        enr_features = np.reshape(np.array(enr).ravel(), (1, len(np.array(con).ravel())))
        dis_features = np.reshape(np.array(dis).ravel(), (1, len(np.array(con).ravel())))
        hom_features = np.reshape(np.array(hom).ravel(), (1, len(np.array(con).ravel())))

        features = np.concatenate((con_features, enr_features, dis_features, hom_features), axis=1)
        features_list.append(features)

    # Convert to DataFrame
    df = pd.DataFrame(np.concatenate(features_list))  # No need for specifying columns
    df['Label'] = labels  # Add label column
    
    return df
```

```python
# Loading image file
data_dir = os.path.expanduser('dataset')
class_names = ['nocots', 'cots']  # Define class names explicitly
image_files = []
labels = []
# Assign labels based on directory names
for class_name in class_names:
    class_dir = os.path.join(data_dir, class_name)
    for file in os.listdir(class_dir):
        if file.endswith('.jpg'):
            image_files.append(os.path.join(class_dir, file))
            if class_name == 'nocots':
                labels.append(1)  # Assign 1 for 'nocots'
            elif class_name == 'cots':
                labels.append(0)  # Assign 0 for 'cots'
```

```python
print(labels)
```

```python
# Extract features for each image with labels
dataframes_list = []
for i in range(len(image_files)):
    features_df = extract_features_from_images([image_files[i]], [labels[i]])
    dataframes_list.append(features_df)
```

```python
# Concatenate DataFrames
final_df = pd.concat(dataframes_list, ignore_index=True)

# Add labels directly to the DataFrame
final_df['Label'] = labels
```

```python
# Save DataFrame to CSV
final_df.to_csv('feature_set.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
# Separate features and labels
X = features_db.drop('Label', axis=1)  # Features
y = features_db['Label']  # Labels

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

# Apply PCA to reduce dimensionality
n_components = 50  # Set the desired number of components (adjust as needed)
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_pca, y_train)
rf_predictions = rf_model.predict(X_test_pca)
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("Random Forest Accuracy with PCA:", rf_accuracy)

# Logistic Regression
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train_pca, y_train)
lr_predictions = lr_model.predict(X_test_pca)
lr_accuracy = accuracy_score(y_test, lr_predictions)
print("Logistic Regression Accuracy with PCA:", lr_accuracy)

# Support Vector Machine
svm_model = SVC(random_state=42)
svm_model.fit(X_train_pca, y_train)
svm_predictions = svm_model.predict(X_test_pca)
svm_accuracy = accuracy_score(y_test, svm_predictions)
print("SVM Accuracy with PCA:", svm_accuracy)

# XGBoost
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(X_train_pca, y_train)
xgb_predictions = xgb_model.predict(X_test_pca)
xgb_accuracy = accuracy_score(y_test, xgb_predictions)
print("XGBoost Accuracy with PCA:", xgb_accuracy)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate accuracy
xgb_accuracy = accuracy_score(y_test, xgb_predictions)

# Calculate precision, recall, and F1 score for XGBoost model
xgb_precision = precision_score(y_test, xgb_predictions)
xgb_recall = recall_score(y_test, xgb_predictions)
xgb_f1 = f1_score(y_test, xgb_predictions)

print("XGBoost Accuracy with PCA:", xgb_accuracy)
print("XGBoost Precision with PCA:", xgb_precision)
print("XGBoost Recall with PCA:", xgb_recall)
print("XGBoost F1 Score with PCA:", xgb_f1)
```
